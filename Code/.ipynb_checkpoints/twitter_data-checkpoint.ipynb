# Topic Modeling

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from wordcloud import WordCloud
import gensim
from gensim.utils import simple_preprocess
import gensim.corpora as corpora
from pprint import pprint
import pyLDAvis.gensim_models as gensimvis
import pickle 
import pyLDAvis
import os
```

# Word Cloud --all

```python
stop_words = stopwords.words('english')
stop_words.extend(stopwords.words('portuguese'))
stop_words.extend(stopwords.words('french'))
stop_words.extend(stopwords.words('german'))
```

```python
long_string = ','.join(list(tweets_df1_topic['FilteredContent'].values))
wordcloud = WordCloud(width = 800, height = 400,background_color="white", max_words=1000, contour_width=5, contour_color='steelblue', stopwords = stop_words)
wordcloud.generate(long_string)
wordcloud.to_image()
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))

def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
```

```python
data = tweets_df1_topic.FilteredContent.values.tolist()
data_words = list(sent_to_words(data))

# remove stop words
data_words = remove_stopwords(data_words)

print(data_words)
```

```python


# Create Dictionary
id2word = corpora.Dictionary(data_words)

# Create Corpus
texts = data_words

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus)
```

```python

# number of topics
num_topics = 3

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)

# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
pyLDAvis.enable_notebook()

LDAvis_data_filepath = os.path.join('../Data/Topic'+str(num_topics))


if 1 == 1:
    LDAvis_prepared = gensimvis.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)


with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)

pyLDAvis.save_html(LDAvis_prepared, '../Data/Topic'+ str(num_topics) +'.html')

LDAvis_prepared
```

```python
# Removing empty tweets
tweets_df1_sentiment = tweets_df1_sentiment[tweets_df1_sentiment['Content'] != '']
tweets_df1_sentiment = tweets_df1_sentiment.reset_index()
```

```python
import flair
```

```python
flair_sentiment = flair.models.TextClassifier.load('en-sentiment')
```

```python
sentiment = []
for index, row in tweets_df1_sentiment.iterrows():  
    s = flair.data.Sentence(str(row["ContentWithoutURL"]))
    flair_sentiment.predict(s)    
    total_sentiment = s.labels
    sentiment.append(str(total_sentiment))
```

```python
sentiment = pd.Series(sentiment)
```

```python
sentiment_df=pd.DataFrame(sentiment,columns=['Sentiment'])
```

```python
sentiment_df[['Sentiment', 'Score']] = sentiment_df['Sentiment'].str.split(' ', expand=True)
```

```python
sentiment_df
```

```python
sentiment_df['Sentiment'] = sentiment_df['Sentiment'].map(lambda x: x.lstrip('['))
```

```python
sentiment_df['Score'] = sentiment_df['Score'].map(lambda x: x.lstrip('(').rstrip(')]'))
```

```python
sentiment_df['Score'] = pd.to_numeric(sentiment_df["Score"])
```

```python
sentiment_df
```

```python
tweets_df1_sentiment=tweets_df1_sentiment.join(sentiment_df, how='outer')
```

```python
tweets_df1_sentiment
```

```python
tweets_df1_sentiment_pos = tweets_df1_sentiment[tweets_df1_sentiment['Sentiment'] == 'POSITIVE'].reset_index()
tweets_df1_sentiment_pos
```

```python
tweets_df1_sentiment_pos = tweets_df1_sentiment_pos.groupby(by='Category')['Sentiment'].value_counts().reset_index(name='count')
tweets_df1_sentiment_pos
```

```python
tweets_df1_sentiment_neg = tweets_df1_sentiment[tweets_df1_sentiment['Sentiment'] == 'NEGATIVE']
tweets_df1_sentiment_neg
```

```python
tweets_df1_sentiment_neg = tweets_df1_sentiment_neg.groupby(by='Category')['Sentiment'].value_counts().reset_index(name='count')
tweets_df1_sentiment_neg
```

```python
frames = [tweets_df1_sentiment_pos, tweets_df1_sentiment_neg]
result = pd.concat(frames)
```
